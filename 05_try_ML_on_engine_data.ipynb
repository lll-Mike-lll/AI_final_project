```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('engine_data_SI_unit.csv')
df.head()
```

```python
df.shape
```

```python
x_rpm = df['rpm'].to_numpy()
x_rpm.reshape(len(x_rpm),1)
```

```python
target_fuel = df['fuel'].to_numpy()
target_fuel.reshape(len(target_fuel),1)
```

```python
x_power = df['power'].to_numpy()
x_power.reshape(len(x_power),1)
```

```python
x_torque = df['torque'].to_numpy()
x_torque.reshape(len(x_torque),1)
```

```python
inputs = np.column_stack((x_rpm,x_power,x_torque))
```

```python
init_range = 1

weights = np.random.uniform(low=-init_range, high=init_range, size=(3, 1))
biases = np.random.uniform(low=-init_range, high=init_range, size=1)
print (weights)
print (biases)
```

```python
weights = [[0.07],[-2.7],[0.17]]
```

```python
learning_rate = 0.001
```

```python
for i in range (10):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - target_fuel
    loss = np.sum(deltas ** 2) / 2 / 2000
    print (loss)
    deltas_scaled = deltas / 2000
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

```python
print (weights, biases)
```

```python
plt.plot(outputs,target_fuel)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
