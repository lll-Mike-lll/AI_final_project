```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn import linear_model
sns.set()
```

```python
df = pd.read_csv('engine_data_SI_unit.csv')
df.head()
```

```python
y = df['fuel']
x1 = df[['rpm','power','torque']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['rpm','power','torque']],df.fuel)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2342.529,55.2874,226.1839]])
```
